```python
df = pd.read_excel('Bitacora_extrusion_limpia.xlsx')
```

```python
df.head(5)
```

```python
dfFecha = df[['N° DE PEDIDO','FECHA']]
d = dfFecha.groupby(['N° DE PEDIDO']).count()
d.hist(bins=150,figsize=(15,8),alpha=0.75)
```

```python
d.plot.bar(figsize=(15,8))
plt.xticks(horizontalalignment="center")
plt.xlim(0,50)
plt.grid()
```

```python
d2 = dfFecha.groupby(['FECHA']).count()
d2.hist(bins=145,figsize=(15,8),alpha=0.75,color='r')
```

```python
d2.describe()
```

```python
d2.plot.bar(figsize=(15,8),color='r')
plt.xticks(horizontalalignment="center")
plt.xlim(0,50)
plt.grid()
```
